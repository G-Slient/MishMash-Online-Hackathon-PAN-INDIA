# This is a Script to Aggregate the Train data from DayWise to PeriodWise

# Import Libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import math
```

```python
# Import the dataset 
train = pd.read_csv("data/train.csv")
test = pd.read_csv("data/test.csv")

train.shape
```

```python
train.head()
```

### As this train data is daily-wise lets convert this into period-wise by aggregating : 
        
        - 1 period  ==  4 weeks  ==  28 days

```python
# Calculate the Period number
train['Period'] = train['Day'].apply(lambda x: math.floor(x/28))
```

```python
# Aggregation using mean
agg = pd.DataFrame(train.groupby(['Period']).apply(np.mean))
agg.drop(['Day', 'Period'], axis=1, inplace=True)
agg.reset_index(inplace=True)
agg.to_csv("train_period_wise_mean_aggregated.csv", index=False)
agg.head()
```

```python
# Aggregation using sum
agg = pd.DataFrame(train.groupby(['Period']).sum())
agg.drop(['Day'], axis=1, inplace=True)
agg.reset_index(inplace=True)
agg.to_csv("train_period_wise_sum_aggregated.csv", index=False)
agg.head()
```

- Aggregation using mean data is more similar to the test data than the aggregated using sum data. 
- So, we are going to use the aggregated using mean data for further predictions/Forecasting.
